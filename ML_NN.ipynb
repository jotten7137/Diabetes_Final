```python
#Dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#sklearn dependencies
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Read diabetes csv
diabetes_data = pd.read_csv('diabetes_clean_data.csv')
diabetes_data.head()
```

```python
#Read nutrition csv
nutrition_data = pd.read_csv('dietary_clean_data.csv')
nutrition_data.head()
```

```python
# #Read demographic csv
# demographic_data = pd.read_csv('demograph_df.csv')
# demographic_data.head()
```

```python
#Pandas get_dummies to convert categorical data
diabetes_data = pd.get_dummies(diabetes_data)
nutrition_data = pd.get_dummies(nutrition_data)
#demographic_data = pd.get_dummies(demographic_data)
```

```python
#Merge
diabetes_nutr = diabetes_data.merge(nutrition_data,how='outer',on=None)
diabetes_nutr.head()
```

```python
diabetes_nutr.dtypes
```

```python
diabetes_nutr= diabetes_nutr.apply(np.int64)
```

```python
diabetes_nutr.dtypes
```

```python
#Variables, target & shape
X = diabetes_nutr[['Overweight','Age','PoorDiet','Race','LackOfPhyAct','HighBP','HighBS','HighChol']] #Variable
y = diabetes_nutr['EvrToldPreDia'].values.reshape(-1,1) #test
print (X.shape, y.shape)
```

```python
X = diabetes_nutr[['ALCOHOL','B12_ADDED','CAFFEINE','CARBOHYDRATES','CHOLESTEROL','FIBER','IRON','CALORIES','FOODCOMPONENT','MAGNESIUM','MONOSATURATED_FAT','POTASSIUM','PROTEIN','SATURATED_FAT','SODIUM',]] #Variable
y = diabetes_nutr['EvrToldPreDia'].values.reshape(-1,1) #test
print (X.shape, y.shape)
```

```python
#Training and Testing
#X=Variable & y=target
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=50)
```

```python
#Create a StandardScater model and fit training
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#encode the labels
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
y_train_cat
```

```python
#Sequential model
model = Sequential()
```

```python
#NN - Input & Hidden Layers
number_inputs = len(X)
number_hidden_nodes = 4
model.add(Dense(units=number_hidden_nodes, activation='relu',input_dim=number_inputs))
```

```python
#NN - Output
number_classes = 2
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
#Model Summary
model.summary()
```

```python
#
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    y_train_cat,
    epochs=1000,
    shuffle=True,
    verbose=5
)
```
